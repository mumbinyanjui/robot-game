```python
#%pip install pygame

import pygame
import random
 
# Initialize pygame
pygame.init()
 
# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Coin Rain Adventure")
 
# Load images
robot_img = pygame.image.load("robot.png")
coin_img = pygame.image.load("coin.png")
monster_img = pygame.image.load("monster.png")
 
# Scale images
robot_img = pygame.transform.scale(robot_img, (50, 50))
coin_img = pygame.transform.scale(coin_img, (30, 30))
monster_img = pygame.transform.scale(monster_img, (50, 50))
 
# Player settings
player_x, player_y = WIDTH // 2, HEIGHT - 60  # Position the player at the bottom
player_speed = 7
score = 0
 
# Coin settings
coins = []
for _ in range(5):  # Start with 5 coins
    coins.append([random.randint(0, WIDTH - 30), random.randint(-600, -30)])
 
# Monster settings
monsters = []
for _ in range(3):  # Start with 3 monsters
    monsters.append([random.randint(0, WIDTH - 50), random.randint(-600, -50)])
 
# Function to draw objects on screen
def draw_objects():
    screen.fill((0, 0, 0))  # Clear the screen
    screen.blit(robot_img, (player_x, player_y))
    for coin in coins:
        screen.blit(coin_img, (coin[0], coin[1]))
    for monster in monsters:
        screen.blit(monster_img, (monster[0], monster[1]))
 
# Function to handle player movement
def handle_player_movement():
    global player_x
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_x > 0:
        player_x -= player_speed
    if keys[pygame.K_RIGHT] and player_x < WIDTH - 50:
        player_x += player_speed
 
# Function to update coin and monster positions and check for collisions
def update_positions():
    global score, running
 
    # Update coin positions and check for collection
    for coin in coins:
        coin[1] += 5  # Coins fall down the screen
        if coin[1] > HEIGHT:
            coin[1] = random.randint(-600, -30)
            coin[0] = random.randint(0, WIDTH - 30)
        
        # Check collision with player
        if abs(player_x - coin[0]) < 40 and abs(player_y - coin[1]) < 40:
            score += 1
            coin[1] = random.randint(-600, -30)
            coin[0] = random.randint(0, WIDTH - 30)
 
    # Update monster positions and check for collision with player
    for monster in monsters:
        monster[1] += 7  # Monsters fall faster
        if monster[1] > HEIGHT:
            monster[1] = random.randint(-600, -50)
            monster[0] = random.randint(0, WIDTH - 50)
 
        # Check collision with player
        if abs(player_x - monster[0]) < 40 and abs(player_y - monster[1]) < 40:
            print("Game Over! You were hit by a monster.")
            running = False
 
# Function to display score
def display_score():
    font = pygame.font.Font(None, 36)
    score_text = font.render(f"Score: {score}", True, (255, 255, 255))
    screen.blit(score_text, (10, 10))
 
# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
 
    handle_player_movement()
    update_positions()
    draw_objects()
    display_score()
    pygame.display.flip()
    pygame.time.delay(30)
 
pygame.quit()
```
